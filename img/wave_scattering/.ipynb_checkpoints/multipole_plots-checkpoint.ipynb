# Imports

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal
import scipy.optimize
```

# Matplotlib params

```python
font_size = 18
mpl.rcParams['mathtext.fontset'] = "custom"
mpl.rcParams['mathtext.rm'] = 'Tex Gyre Pagella Math'
mpl.rcParams['mathtext.it'] = 'Tex Gyre Pagella Math:italic'
mpl.rcParams['mathtext.bf'] = 'Tex Gyre Pagella Math:bold'
mpl.rcParams['font.family'] = 'TeX Gyre Pagella'
plt.rcParams['figure.figsize'] = [10, 8]
```

# Full Data

```python
# Available: 20/40, 20/20 and 00/20
l = "2"
m = "0"
r = "40"

file_path = "/home/lucas/EinsteinToolkit/Cactus/exe/"

file_name = "KG_Phi_" + l + m + "_r" + r + ".0_07_04_2022_20:35:54.ascii"

file_name = file_path + file_name
```

```python
data = np.loadtxt(file_name, delimiter=" ")
data_x = data[:, 0]
data_y = data[:, 1]
```

# Time sliced data

```python
delim_plot_start_time = 67.2
delim_plot_plot_end_time = 136.5

time_step = data_x[1] - data_x[0]
delim_plot_start_idx = int(np.ceil(delim_plot_start_time/time_step)) + 1
delim_plot_end_idx = int(np.ceil(delim_plot_plot_end_time/time_step)) + 1

time_range = slice(delim_plot_start_idx, delim_plot_end_idx)
data_x_sliced = data_x[time_range]
data_y_sliced = data_y[time_range]
```

# Plots

```python
plt.close("all")

fig, axes = plt.subplots(nrows=2, ncols=2)

# Full plots

axes[0,0].set_xlabel(r"$t$", fontsize=font_size)
axes[0,0].set_ylabel(fr"$\Phi_{{{l}{m}}}({r},t)$", fontsize=font_size)
axes[0,0].tick_params(axis="both", which="major", labelsize=font_size)
axes[0,0].plot(data_x, data_y, color="black", linewidth=2)

axes[0,1].set_xlabel(r"$t$", fontsize=font_size)
axes[0,1].set_ylabel(fr"$\Phi_{{{l}{m}}}({r},t)$", fontsize=font_size)
axes[0,1].tick_params(axis="both", which="major", labelsize=font_size)
axes[0,1].plot(data_x, np.log(np.abs(data_y)), color="black", linewidth=2)

# Time sliced plots

axes[1,0].set_xlabel(r"$t$", fontsize=font_size)
axes[1,0].set_ylabel(fr"$\Phi_{{{l}{m}}}({r},t)$", fontsize=font_size)
axes[1,0].tick_params(axis="both", which="major", labelsize=font_size)
axes[1,0].plot(data_x_sliced, data_y_sliced, color="black", linewidth=2)

axes[1,1].set_xlabel(r"$t$", fontsize=font_size)
axes[1,1].set_ylabel(fr"$\Phi_{{{l}{m}}}({r},t)$", fontsize=font_size)
axes[1,1].tick_params(axis="both", which="major", labelsize=font_size)
axes[1,1].plot(data_x_sliced, np.log(np.abs(data_y_sliced)), color="black", linewidth=2)

fig.tight_layout()
fig.savefig("multipolar_plots.pdf")
```

# PSD plot

```python
plt.close("all")

(f,s) = scipy.signal.periodogram(data_y, 1/time_step)

max_psd_idx = s.argmax(axis=0)
max_psd = s[max_psd_idx]
max_f = f[max_psd_idx]

plt.xlabel(r"$f$", fontsize=font_size)
plt.ylabel("PSD", fontsize=font_size)
plt.tick_params(axis="both", which="major", labelsize=font_size)
plt.plot(f, s, color="black", linewidth=2)

plt.text(0.9, 0.001, fr"Max. $f = {{{max_f}}}$", fontsize=font_size)
plt.text(0.9, 0.00094, fr"Max. $PSD = {{{max_psd}}}$", fontsize=font_size)

plt.tight_layout()
plt.savefig("psd_plot.pdf")
```

# Fit

```python
def qnm_model(t, A, wr, wi, phi):
    return A * np.exp(-wi * t) * np.cos(wr * t - phi)

fit_data = scipy.optimize.curve_fit(qnm_model, data_x_sliced, data_y_sliced, p0=[0.06, 0.1, 0.1, 0.1])
data_y_fit = qnm_model(data_x_sliced, *fit_data[0])
data_y_residual = np.abs(data_y_sliced - data_y_fit)

plt.close("all")

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_xlabel(r"$t$", fontsize=font_size)
axes[0].set_ylabel(fr"$\Phi_{{{l}{m}}}({r},t)$", fontsize=font_size)
axes[0].tick_params(axis="both", which="major", labelsize=font_size)
axes[0].plot(data_x_sliced, data_y_sliced, color="black", linewidth=2, label="Data")
axes[0].plot(data_x_sliced, data_y_fit, color="red", linewidth=2, label="Fit")
axes[0].legend(loc="upper right")
axes[0].text(88, 0.05, fr"$\omega_r = {{{fit_data[0][1]}}}$", fontsize=font_size - 3)
axes[0].text(88, 0.046, fr"$\omega_i = {{{fit_data[0][2]}}}$", fontsize=font_size - 3)

axes[1].set_xlabel(r"$t$", fontsize=font_size)
axes[1].set_ylabel("Fit Residual", fontsize=font_size)
axes[1].tick_params(axis="both", which="major", labelsize=font_size)
axes[1].plot(data_x_sliced, data_y_residual, color="black", linewidth=2)

fig.tight_layout()
fig.savefig("fit_plot.pdf")
```
